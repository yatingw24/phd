```python
import pandas as pd

# Define the URL
url = "https://www.phdstipends.com/csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(url)

# Save it locally as a CSV file
df.to_csv("phd_stipends.csv", index=False)

# Display the first few rows
df
```

```python
df = df.dropna(subset=['Department'])
df['Overall Pay'] = df['Overall Pay'].astype(str).str.replace('[$,]', '', regex=True).astype(float)
```

```python
df['Overall Pay'] = pd.to_numeric(df['Overall Pay'], errors='coerce')

# Drop rows where "Overall Pay" is negative or NaN
df = df[df['Overall Pay'] >= 0]
```

```python
# drop rows where Department is marked by non-English characters

df = df[df["Department"].str.match(r"^[A-Za-z\s\-&]+$", na=False)]
```

```python
df["Department"].value_counts()
```

# create a new column that categorizes majors into Fields of Study

1. categorization
2. dropped any "NA" values and unspecified departments

```python
df.loc[df["Department"].str.contains("clinic|molepatho|engi|bio|phy|ocea|mistry|anatomy|cs|comput|math|stat|neu|quan|gene|immu|tech|pharm|mechan|micro|geo|data|medical|material|medicine|eco|nursing|robot|earth|electri|health|animal|information|epide", case=False, na=False), "Field"] = "STEM"
df.loc[df["Department"].str.contains("account|manage|MBA|business|finance|marketing|administration", case=False, na=False), "Field"] = "Business"
df.loc[df["Department"].str.contains("international|psych|life|nutrition|edu|econ|gover|public|policy|linguistics|soci|comm|politic|development|social|food|organizational|couns|crim|city", case=False, na=False), "Field"] = "Social Science"
df.loc[df["Department"].str.contains("humanities|spanish|german|english|music|history|anthropology|philosophy|american|literature|theat|Studies|religion|archi|culture|language", case=False, na=True), "Field"] = "Humanities"
```

```python
# Replace the string 'NaN' with actual NaN (using numpy)
import numpy as np
df['Overall Pay'] = df['Overall Pay'].replace("NaN", np.nan)

# Now drop rows with NaN values in 'Overall Pay'
df = df.dropna(subset=["Overall Pay"])
df = df[(df['Overall Pay'] >= 15000) & (df['Overall Pay'] <= 90000)]
```

```python

df['University'] = df['University'].replace("NaN", np.nan)

# Now drop rows with NaN values in 'Overall Pay'
df = df.dropna(subset=["University"])
```
