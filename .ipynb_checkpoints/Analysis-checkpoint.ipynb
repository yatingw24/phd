```python
import pandas as pd

# Define the URL
url = "https://www.phdstipends.com/csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(url)

# Save it locally as a CSV file
df.to_csv("phd_stipends.csv", index=False)

# Display the first few rows
df
```

```python
df_recent = df[(df['Academic Year'] == '2023-2024') | (df['Academic Year'] == '2024-2025') | (df['Academic Year'] == '2022-2023')]
```

```python
df_recent[df_recent['Department'] == 'Physics'].head(40)

```

```python
df_recent = df_recent.dropna(subset=['Department'])
df_recent['Overall Pay'] = df_recent['Overall Pay'].astype(str).str.replace('[$,]', '', regex=True).astype(float)
```

```python
df_recent['Overall Pay'] = pd.to_numeric(df_recent['Overall Pay'], errors='coerce')

# Drop rows where "Overall Pay" is negative or NaN
df_recent = df_recent[df_recent['Overall Pay'] >= 0]
```
